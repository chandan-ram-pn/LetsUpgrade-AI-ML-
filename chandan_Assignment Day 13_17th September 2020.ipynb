```python
import matplotlib.pyplot as mpp
x=np.array(credit_cards['Time'])
y=np.array(credit_cards['Amount'])
mpp.plot(x,y,'k')
mpp.xlabel('TIME')
mpp.ylabel("AMOUNT")
mpp.show()
```

```python
import matplotlib.pyplot as mpp
x=np.array(credit_cards.head(15)['Time'])
y=np.array(credit_cards.head(15)['Amount'])
mpp.bar(x,y)
mpp.xlabel('TIME')
mpp.ylabel("AMOUNT")
mpp.title('time vs amount')
mpp.show()
```

```python
import matplotlib.pyplot as mpp
x=np.array(credit_cards.head(15)['Time'])
y=np.array(credit_cards.head(15)['Amount'])
mpp.hist(x)
mpp.xlabel('TIME')
mpp.ylabel("AMOUNT")
mpp.title('time vs amount')
mpp.show()
```
